```python
%%bash
pwd
source /content/venv310/bin/activate
touch .env                            # Hydra / dotenv happy

# ─── choose slice to run ────────────────────────────────────────────────
export SAMPLING_MIN=15                # 5, 15, 30, 45, 60
export NUM_NODES=8                    # 8, 16, 25
# ────────────────────────────────────────────────────────────────────────

/content/venv310/bin/python - <<'PY'
import os, sys, pandas as pd, numpy as np, matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from pathlib import Path
from gluonts.dataset.pandas import PandasDataset
from uni2ts.model.moirai import MoiraiForecast, MoiraiModule
from uni2ts.eval_util.plot import plot_single

# ─── read env-vars ───────────────────────────────────────────────────────
SAMPLING_MIN = int(os.environ["SAMPLING_MIN"])
NUM_NODES    = int(os.environ["NUM_NODES"])

# TABLE constants (in hours) ---------------------------------------------
WINDOW_HR   = 8
HORIZON_HR  = 4
STRIDE_STEPS= 1
BATCH_SIZE  = 64

# AUTO-SCALED step counts --------------------------------------------------
CTX_STEPS   = WINDOW_HR  * 60 // SAMPLING_MIN     # 8 h in steps
PRED_STEPS  = HORIZON_HR * 60 // SAMPLING_MIN     # 4 h in steps

print(f"sampling={SAMPLING_MIN} min | context={CTX_STEPS} | horizon={PRED_STEPS}")

csv = Path(f"/content/Dataset_perSampling_pernodeConfig/"
           f"temperature_data_{SAMPLING_MIN}min_{NUM_NODES}.csv")
df  = pd.read_csv(csv, index_col=0, parse_dates=True)

# 5-day / 2-day / 2-day split (timestamps) --------------------------------
df_train = df.loc["2018-11-01":"2018-11-06"]
df_test  = df.loc["2018-11-08":"2018-11-10"]

train_ds = PandasDataset(dict(df_train))
test_ds  = PandasDataset(dict(df_test))

# 3) Zero-shot Moirai inference
model = MoiraiForecast(
    module=MoiraiModule.from_pretrained("Salesforce/moirai-1.0-R-small"),
    prediction_length=PRED_STEPS,
    context_length=CTX_STEPS,
    patch_size="auto",
    num_samples=100,
    target_dim=1,
    feat_dynamic_real_dim=train_ds.num_feat_dynamic_real,
    past_feat_dynamic_real_dim=train_ds.num_past_feat_dynamic_real,

)
pred = model.create_predictor(batch_size=BATCH_SIZE)

forecast = next(iter(pred.predict(test_ds)))
rec      = next(iter(test_ds))
target   = rec["target"]

context = target[-(CTX_STEPS+PRED_STEPS):-PRED_STEPS]
truth   = target[-PRED_STEPS:]

freq      = pd.Timedelta(minutes=SAMPLING_MIN)
ctx_start = rec["start"] + freq*(len(target)-len(context))
lbl_start = ctx_start + freq*len(context)


from sklearn.metrics import mean_absolute_error, mean_squared_error

# Get forecast mean prediction
pred_mean = forecast.mean

# Convert both to numpy arrays
truth_arr = np.array(truth)
pred_arr  = np.array(pred_mean)

# Evaluation metrics
mae  = mean_absolute_error(truth_arr, pred_arr)
mape = np.mean(np.abs((truth_arr - pred_arr) / truth_arr)) * 100
rmse = np.sqrt(mean_squared_error(truth_arr, pred_arr))

print(f"MAE  = {mae:.3f}")
print(f"MAPE = {mape:.3f}%")
print(f"RMSE = {rmse:.3f}")


plot_single({"start": ctx_start, "target": context},
            {"start": lbl_start, "target": truth},
            forecast,
            context_length=CTX_STEPS,
            name="pred", show_label=True)

out = f"/content/pred_{SAMPLING_MIN}min_{NUM_NODES}nodes.png"
plt.savefig(out, bbox_inches="tight"); plt.close()
print("✓ saved →", out)
PY
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the metrics_log.csv file (adjust the path to your CSV file)
df = pd.read_csv('metrics_log.csv')

# Prepare figure with 3 plots (MAE, MAPE, RMSE)
fig, axs = plt.subplots(1, 3, figsize=(18, 8))  # Increased figure size for better readability

# Define the sampling rates (x-axis values)
sampling_rates = sorted(df['SAMPLING_MIN'].unique())

# Define the node counts (x-axis values)
node_counts = sorted(df['NUM_NODES'].unique())

# Define colors for each sampling rate (5min, 15min, 30min, etc.)
colors = ['b', 'g', 'r', 'c', 'm']  # Blue, Green, Red, Cyan, Magenta

# Plot MAE
for i, rate in enumerate(sampling_rates):
    rate_data = df[df['SAMPLING_MIN'] == rate]
    axs[0].plot(rate_data['NUM_NODES'], rate_data['MAE'], marker='o', color=colors[i], label=f'{rate} min', linewidth=2)  # Increase line width
axs[0].set_title('MAE vs Number of Nodes', fontsize=16)  # Increase title font size
axs[0].set_xlabel('Number of Nodes', fontsize=14)  # Increase x-axis label font size
axs[0].set_ylabel('MAE', fontsize=14)  # Increase y-axis label font size
axs[0].legend(title="Sampling Rates", fontsize=12)  # Increase legend font size

# Plot MAPE
for i, rate in enumerate(sampling_rates):
    rate_data = df[df['SAMPLING_MIN'] == rate]
    axs[1].plot(rate_data['NUM_NODES'], rate_data['MAPE'], marker='o', color=colors[i], label=f'{rate} min', linewidth=2)  # Increase line width
axs[1].set_title('MAPE vs Number of Nodes', fontsize=16)  # Increase title font size
axs[1].set_xlabel('Number of Nodes', fontsize=14)  # Increase x-axis label font size
axs[1].set_ylabel('MAPE (%)', fontsize=14)  # Increase y-axis label font size
axs[1].legend(title="Sampling Rates", fontsize=12)  # Increase legend font size

# Plot RMSE
for i, rate in enumerate(sampling_rates):
    rate_data = df[df['SAMPLING_MIN'] == rate]
    axs[2].plot(rate_data['NUM_NODES'], rate_data['RMSE'], marker='o', color=colors[i], label=f'{rate} min', linewidth=2)  # Increase line width
axs[2].set_title('RMSE vs Number of Nodes', fontsize=16)  # Increase title font size
axs[2].set_xlabel('Number of Nodes', fontsize=14)  # Increase x-axis label font size
axs[2].set_ylabel('RMSE', fontsize=14)  # Increase y-axis label font size
axs[2].legend(title="Sampling Rates", fontsize=12)  # Increase legend font size

# Adjust layout for better spacing between subplots
plt.tight_layout()

# Save the plot to a file (optional)
plt.savefig("metrics_plot_nodes_vs_error.png", dpi=1200)  # Increase DPI for better quality

# Display the plot
plt.show()
```
